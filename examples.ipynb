```python
import leukemiadrugscreen as drugscreen

# loading example DeviceStack
example = drugscreen.DeviceStack.from_folder("leukemiadrugscreen\example_data\JURKATS DVP")
```

## Accessing drug fit parameters
Main hill parameters: accessed using Emax, E0, C, or h:

```python
print(f"Emax of drug 1: {example.singledrugfits[0].Emax:.3f}")
print(f"EC50 of drug 2: {example.singledrugfits[1].C:.3f} {example.dose_units[1]}")
```

Defining a region based on a ternary point, fitting a hill curve for viability, and getting the AUC up to the EC50:

```python
ternary_point = (0.5, 0.5, 0) # 50% of drug A, 50% of drug B, and 0% of drug C
region = drugscreen.Region.from_point(ternary_point, example)
hillfit = region.fit_hill()
hillfit.plot_curve()
print(f"AUC of curve at EC50: {hillfit.get_AUC(EC = 50)}")
print(f"AUC of curve at EC30: {hillfit.get_AUC(EC = 30)}")
```
