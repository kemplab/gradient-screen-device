```python
import matplotlib.pyplot as plt
from drugfit import Hill
import numpy as np

potency = Hill(E0 = 1, Emax = 0.4, h = 4, C = 1)
efficacy = Hill(E0 = 1, Emax = 0.1, h = 4, C = 2)

fig = plt.figure("potencyefficacy", figsize = (4, 3), dpi = 150)
ax = fig.subplots()
xplot = np.linspace(0, 4, 100)

# plots
ax.plot(xplot, potency.E(xplot), linewidth = 3, label = "Increased Potency", c = "slateblue")
ax.plot(xplot, efficacy.E(xplot), linewidth = 3, label ="Increased Effect", c = "coral")

#ax.legend(loc = "upper right", fontsize = 12)

#plt.xticks([])
#plt.yticks([])
#ax.xaxis.set_ticklabels([])
ax.set_xticks([])
ax.set_yticks([])
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)

#ax.set_xlabel("Drug Dose", fontsize = 14)
#ax.set_ylabel("Drug Effect", fontsize = 14)

plt.show()
```

```python
import matplotlib.pyplot as plt
from drugfit import Hill
import numpy as np

potency1 = Hill(E0 = 1, Emax = 0.2, h = 4, C = 1.5)
potency2 = Hill(E0 = 1, Emax = 0.2, h = 4, C = 1)
potency3 = Hill(E0 = 1, Emax = 0.2, h = 4, C = 0.5)


fig = plt.figure("potency", figsize = (3, 2), dpi = 150)
ax = fig.subplots()
xplot = np.linspace(0, 4, 100)

# plots
ax.plot(xplot, potency1.E(xplot), linewidth = 3, c = "slateblue", alpha = 0.6)
ax.plot(xplot, potency2.E(xplot), linewidth = 3, c = "slateblue", alpha = 0.8)
ax.plot(xplot, potency3.E(xplot), linewidth = 3, c = "slateblue", alpha = 1.0)


ax.set_xticks([])
ax.set_yticks([])
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)
ax.set_ylim(0, 1)

ax.set_xlabel("Drug Dose", fontsize = 14)
ax.set_ylabel("Cell Viability", fontsize = 14)

plt.show()
```
